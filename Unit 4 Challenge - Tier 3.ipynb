```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

```python
# taking the annual averages for the past 20 years of date, i.e. since 2002
london_properties_annual = london_properties_tall.reset_index()
london_properties_annual['year'] = london_properties_annual['date'].dt.year
london_properties_annual = (
    london_properties_annual
    [['year', 'borough', 'average_price']]
    .query('year >= 2002')
    .groupby(['borough', 'year'], as_index=False)
    .mean()
)
london_properties_annual.head()
```

```python
# extract the 2002 values as a baseline
baseline_values = london_properties_annual.loc[
    london_properties_annual['year'] == 2002,
    ['borough', 'average_price']
]
# set borough as an index for easier access in calculation
baseline_values = baseline_values.set_index('borough')

# calculate average prices indexed such that the 2002 value is 100
london_properties_annual['indexed_price'] = london_properties_annual.apply(
    lambda row: row['average_price'] / baseline_values.loc[row['borough']] * 100,
    axis = 1
)
# print the boroughs with the highest, median, and lowest increase
print(london_properties_annual.query('year == 2022').sort_values('indexed_price', ascending=False).iloc[[0,1,15,16,-2,-1]])
```

Is there any association between the initial level of prices and subsequent growth? No, if anything there might be a little bit of convergence.

```python
plt.clf()
plt.scatter(
    baseline_values,
    london_properties_annual.query('year == 2022').sort_values('borough')['indexed_price']
)

plt.show()
```

To test this, we can look at the coefficient of variation. This shows that over the last 20 years, there has been neither convergence nor divergence, though interestingly the trend over time is an inverted-u.

```python
london_properties_annual.groupby('year').agg({'average_price':lambda x: np.std(x)/np.mean(x)})
```

#### Summary

This exercise sought to answer the question of which borough of London saw the greatest increase in property values in the past 20 years. I found that between 2002 and 2022, average property prices increased the most in Hackney, by over 250%. In comparsion, median increase among the boroughs was around 180% times, and the minimum around 150% times. One the one hand, more than doubling in 20 years for even the slowest-rising borough is a quite high rate of property prices inflation. By contrast, the British consumer price inflation index increased under 65% in that same time frame.

On the other hand, nonetheless, the different borough experiences significantly different growth rates. This raises the question of what drove these different rates. From the data available, it is only possible to rule out one hypothesis: it was not the case that the already-expensive boroughs got disproportionately more experience, or vice versa. That is to say, the rate of increase from 2002 to 2022 was not associated with the average price in 2002. Looking at annual patterns rather than just the start and end points suggests a slightly more subtle story, however: the coefficient of variation across boroughs by year shows a distinct inverted-U shape. The dispersion in price among boroughs increased markedly in the 2010s, but by 2022 it had fallen almost to its initial value. This suggests that there has been some "catch up" among the less-pricey boroughs in the last few years. 

#### Data and Methods

The data comes from the London government's Datastore. The original files provide monthly average property values in current pounds sterling for each of London's 32 boroughs, plus the City of London, as well as averages for larger regions.

For the purposes of this analysis, the observations other than the 32 boroughs of London were dropped, and the monthly values were averaged by year for each year from 2002 to 2022. To measure the increase, an index was calculated such that each borough's 2002 average price was 100.  

#### Further analysis

There are limits to what can be ascertained from average values alone. To investigate alternative explanations of the dynamics of housing prices in London, it would be necessary to collect additional data.

For example, do the different increases in housing prices reflect employment opportunities? Assessing this hypothesis would require data on the distribution of job growth within London and about the transportation network, to measure the "nearness" of different boroughs to job opportunitiies. Alternatively, do housing prices respond to the growth of supply in different boroughs? With data on housing construction, it would be possible to estimate whether supply constrains in particularly boroughs increase prices, at least in the short term.
